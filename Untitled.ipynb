```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
import PIL
```

```python
img_path = r"./bot_ocr/test/test.png"
img = cv2.imread(img_path)
h, w, c = img.shape
print(f"image shape| h: {h}, w: {w}, c: {c}")
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 変換前後の対応点を設定
p_original = np.float32([[5, 90], [340, 5], [5,360], [340, 450]])
p_trans = np.float32([[0,0], [524,0], [0,478], [524,478]])
 
# 変換マトリクスと射影変換
M = cv2.getPerspectiveTransform(p_original, p_trans)
i_trans = cv2.warpPerspective(img, M, (524, 478))
```

```python
plt.imshow(cv2.cvtColor(i_trans, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
cv2.rectangle(i_trans, (100, 0), (400, 400), (0, 0, 0), thickness=-1)
plt.imshow(cv2.cvtColor(i_trans, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_path = r"./bot_ocr/iCloud Photos/IMG_2980.PNG"
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
h, w = img.shape
print(f"image shape| w: {w}, h: {h}")
p_original = np.float32([[75, 160], [890, 130], [65, 620], [880, 680]])
p_trans = np.float32([[100, 150], [900, 150], [100,670], [900, 670]])

M = cv2.getPerspectiveTransform(p_original, p_trans)
i_trans = cv2.warpPerspective(img, M, (900, 680), flags=cv2.INTER_CUBIC)

# i_masked = i_trans
i_masked = cv2.rectangle(i_trans, (0, 0), (900, 180), (0), thickness=-1)
i_masked = cv2.rectangle(i_trans, (0, 0), (105, 680), (0), thickness=-1)
# i_masked = cv2.rectangle(i_trans, (135, 0), (900, 680), (0), thickness=-1)
i_masked = cv2.rectangle(i_trans, (135, 0), (570, 680), (0), thickness=-1)

_, i_bin = cv2.threshold(i_masked, 108, 255, cv2.THRESH_BINARY_INV)



fig, ax = plt.subplots(figsize=(12,10))
ax.imshow(i_bin)
plt.show()
cv2.imwrite(img_path+"t.png",  i_bin)
```
